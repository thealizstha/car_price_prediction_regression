```python
fuel = pd.get_dummies(auto['fueltype'], drop_first = True)
fuel
```

```python
auto = pd.concat([auto, fuel], axis = 1)
auto
```

```python
auto = auto.drop(['fueltype'],axis = 1)
auto
```

```python
auto.corr()
```

```python
# Starting machine learning
import seaborn as sns
plt.figure(figsize=(15,5))
sns.heatmap(auto.corr(),annot=True)
plt.show()
```

```python
auto = auto.drop(['gas'],axis = 1)
auto
```

```python
from sklearn.model_selection import train_test_split

# We specify this so that the train and test data set always have the same rows, respectively
df_train, df_test = train_test_split(auto, train_size = 0.85, test_size = 0.15, random_state = 1)
```

```python
from sklearn.linear_model import LinearRegression
X_train = df_train[['horsepower', 'enginesize', 'highwaympg']]

y_train = df_train['price']
# df_test_predict = X_train.iloc[0:1]
lm = LinearRegression()
model = lm.fit(X_train, y_train)

# model.predict(df_test_predict)
```

```python
data = df_test.iloc[10:11]
data
```

```python
# Testing
data = df_test.iloc[10:11]
actual_price = data.pop('price')

# # print(data)
predicted_price = model.predict(data)
print("Actual Price",actual_price)
print("Predicted Price",predicted_price)
```

```python
data = {'horsepower':[140],
        'enginesize':[131],
        "highwaympg":[20],
       }
df = pd.DataFrame(data)
predicted_price = model.predict(df)
print("Actual Price",actual_price)
print("predicted Price",predicted_price)
```
